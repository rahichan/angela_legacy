```python
import requests
from requests.auth import HTTPBasicAuth
import pandas as pd

username = 'apiuser-2d5db65a05c5@apiconnector.com'
password = 'FSG2020!'

endpoint = 'https://r1-api.dotmailer.com'
```

```python
#Get account info
#https://developer.dotdigital.com/reference/account-information

url = endpoint + "/v2/account-info"
query_params = {'format':'text'}

acc_info = requests.request("GET", url, auth=HTTPBasicAuth(username, password), params=query_params)

print(acc_info.status_code)
```

```python
# Get all campaigns
# https://developer.dotdigital.com/reference/campaigns#get-all-campaigns
all_campaigns = []
select = 1000
skip = 0

#endpoint = 'https://r1-api.dotmailer.com'
url = endpoint + "/v2/campaigns"
query_params = {'Select': select, 'Skip': skip}

while True:
    response = requests.request("GET", url, auth=HTTPBasicAuth(username, password), params=query_params)
    for campaign in response:
        all_campaigns.append(campaign)

    if len(all_campaigns) < select:
        # This means there are no more campaigns to fetch.
        break

    skip += select
j = response.json()

campaigns_df = pd.DataFrame.from_dict(j)
campaigns_df.head()
```

```python
#Get one campign summary could be any, I chose that one...
#Get campaign 445040 summary

url = endpoint + "/v2/campaigns/445040/summary"
query_params = {'format':'json'}
#query_params = {'Select': select, 'Skip': skip}

camp_summary = requests.request("GET", url, auth=HTTPBasicAuth(username, password), params=query_params)

camp_445040_summary = pd.DataFrame.from_dict(camp_summary.json(), orient='index')
camp_445040_summary = camp_445040_summary.transpose()
camp_445040_summary
```

```python
camp_445040_summary['campaign_id'] = 445040
```

```python
campaigns_summary = camp_445040_summary.loc[:,:]
campaigns_summary
```

```python
#Get other campigns summaries

for i in campaigns_ids:
    
    url = endpoint + "/v2/campaigns/" + str(i) + "/summary"
    query_params = {'format':'json'}

    camp_summary = requests.request("GET", url, auth=HTTPBasicAuth(username, password), params=query_params)
 
    camp_summ = pd.DataFrame.from_dict(camp_summary.json(), orient='index')
    camp_summ = camp_summ.transpose()
    camp_summ['campaign_id'] = i

    campaigns_summary = campaigns_summary.append(camp_summ)
campaigns_summary.head()
```

```python
campaigns_sum = campaigns_summary.loc[:,['campaign_id', 'dateSent', 'percentageDelivered', 'percentageUniqueOpens',
       'percentageOpens', 'percentageUnsubscribes', 'percentageReplies',
       'percentageHardBounces', 'percentageSoftBounces',
       'percentageUsersClicked', 'percentageClicksToOpens']]
campaigns_sum.dropna(axis=0, inplace=True)
campaigns_sum.drop_duplicates(subset=['campaign_id'], inplace=True)
campaigns_sum.reset_index(drop=True, inplace=True)
campaigns_sum
```
