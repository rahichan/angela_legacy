```python
contacts = pd.read_csv("All-contacts.csv")
contacts.head()
```

```python
berlin_AGU = AGU_backend[["Status", "Shipping Address", "Customer Email"]]
berlin_AGU.head()
```

```python
berlin_AGU["Shipping Address"]
```

```python
berlin_AGU['Place_zipcode'] = None
for i in range(len(berlin_AGU)):
    berlin_AGU['Place_zipcode'][i] = berlin_AGU["Shipping Address"][i][-12:-6]
```

```python
berlin_AGU.head()
```

```python
contacts_berlin = []
for i in range(len(berlin_AGU["Shipping Address"])):
    if berlin_AGU["Status"][i] == "complete":
        if berlin_AGU["Shipping Address"][i][-12:-6] == "Berlin":
            contacts_berlin.append(berlin_AGU["Customer Email"][i])
            
```

```python
len(contacts_berlin)
```

```python
all_berlin = []
for i in range(len(berlin_AGU["Shipping Address"])):
    if berlin_AGU["Shipping Address"][i][-12:-6] == "Berlin":
        all_berlin.append(berlin_AGU["Customer Email"][i])

len(all_berlin)
```

```python
newsletter_berlin = []

for e in contacts_berlin:
    for i in range(len(contacts)):
        if contacts.Email[i] == e:
            newsletter_berlin.append(contacts.Email[i])
```
